```python
countries_df = pd.read_csv("countries.csv")
countries_df
```

```python
# compare the population growth between UK and AUSTRALIA
```

```python
countries_df.country == "United Kingdom"
```

```python
uk = countries_df[countries_df.country == "United Kingdom"]
```

```python
uk
```

```python
aus = countries_df[countries_df.country == "Australia"]
```

```python
aus
```

```python
plt.plot(uk.year, uk.population / 10**6)
plt.plot(aus.year, aus.population / 10**6)
plt.legend(["United Kingdom", "Australia"])
plt.xlabel("Year")
plt.ylabel("Population")
plt.show()
```

```python
uk.population
```

```python
uk.population / uk.population.iloc[0] * 100
```

```python
plt.plot(uk.year, uk.population / uk.population.iloc[0] * 100)
plt.plot(aus.year, aus.population / aus.population.iloc[0] * 100)
plt.legend(["United Kingdom", "Australia"])
plt.xlabel("Year")
plt.ylabel("Population growth (first year = 100%)")
plt.show()
```
